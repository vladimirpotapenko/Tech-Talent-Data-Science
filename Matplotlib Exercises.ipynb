# Ex. 1

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from numpy import random as rd
%matplotlib inline

sales_data = pd.read_csv('company_sales_data.csv')
print(sales_data)

TP_LP, ax = plt.subplots()
ax.plot("month_number", "total_profit", data=sales_data)
ax.set(xlabel="Month Number", ylabel="Total Profit", 
       title="Total Profit by Month", ylim=(150000,450000))
ax.grid(True)
```

# Ex. 2

```python
TP_LP2 = plt.figure()
ax2 = TP_LP2.add_subplot()
ax2.set(facecolor='black')
ax2.plot("month_number", "total_profit", data=sales_data,c='r',
         ls=':',marker='.',lw=3,ms=15, label="Total Profit over Time")
ax2.set(xlabel="Month Number", ylabel="Sold Units Number", 
       title="Total Profit by Month", ylim=(150000,450000))
ax2.grid(True)
ax2.legend(loc='lower right')
```

# Ex. 3

```python
#facecream 	facewash 	toothpaste 	bathingsoap 	shampoo 	moisturizer 	
TP_LP3 = plt.figure()
ax3 = TP_LP3.add_subplot()
ax3.plot("month_number", "facecream", data=sales_data, marker='v',ls=':')
ax3.plot("month_number", "facewash", data=sales_data, marker='v')
ax3.plot("month_number", "toothpaste", data=sales_data,marker="d")
ax3.plot("month_number", "bathingsoap", data=sales_data, marker='D',ls="-.", mfc='pink', mec='yellow')
ax3.plot("month_number", "shampoo", data=sales_data, marker='*')
ax3.plot("month_number", "moisturizer", data=sales_data, marker=".", ms=10, ls=':')
ax3.set(xlabel="Month Number", ylabel="Sold Units Number", title="All Product Sales")
ax3.grid(True)
ax3.legend(loc='upper left')
```

# Ex. 4

```python
plt.style.use('seaborn')
TP_SP = plt.figure()
ax4 = TP_SP.add_subplot()
ax4.scatter("month_number", "toothpaste", data=sales_data, edgecolors='b',c='yellow')
ax4.grid(True,linestyle='-',color='pink')
ax4.set(xlabel='Month Number', ylabel='Toothpaste Sold Units Number',
       title='Toothpaste Sold by Month')
```

# Ex. 5

```python
months = np.arange(len(sales_data.month_number))

width=.35
TP_BP = plt.figure()
ax5 = TP_BP.add_subplot()
fw = ax5.bar(months - width/2, "facewash", width=.35, 
             data=sales_data, label='facewash', color='b')
fc = ax5.bar(months + width/2, "facecream", width=.35, 
             data=sales_data, label='facecream', color='yellow')
ax5.set(xlabel='Months Number',ylabel='Number of Units Sold', title='Exercise 5')
ax5.legend((fw,fc),['facewash','facecream'])
ax5.grid(True)
```

# Ex. 6

```python
x=np.arange(0,20)
y1=x
y2=x**2
y3=x**3
fig_to_save, ax6 = plt.subplots()
ax6.plot(x, y1, 'ro:',
         x, y2, 'b^-.',
         x, y3, 'g*--', ms=10)
ax6.set(xlabel='X', ylabel='Y', title='Y by X')
ax6.grid(color='pink')
ax6.legend(['X','X**2','X**3'], loc='center', 
           frameon=True, facecolor='yellow')

print(fig_to_save.canvas.get_supported_filetypes())

fig_to_save.savefig('vector_figA.svg')
fig_to_save.savefig('raster_figB.png')
```

 I chose a line plot due its classic elegance, and straightforward way in demonstrating variable relationship. Also used for regression, this chart serves a large purpose in data analysis and shouls not be overlooked. With respect to raster or vector, I prefer one or the other depending on the situation. If I need to be able to scale the image to larger sizes, vector would be best as image quality is not degraded in this process due to its foundation in curves and math. If I need more control over color blending I would choose raster.
