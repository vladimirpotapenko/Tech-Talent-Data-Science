# WSP Part 1

On the Monster website: https://www.monster.com/jobs/search?q=junior+data+scientist&where=new+york%2C+ny -- we have 
the individual job title under the h2 tag. Within this tag there are various attributes such as "class," "title," and "name"
among others. If you click on a specific job and view details, then the html is in a div tag, with attributes such as
"prefetch" and "headervariation." Also, the general container of an individual job is also under a div tag with a tabindex and href attributes, with class="results-card". Inside the entire container is also a button tag and a span tag.
